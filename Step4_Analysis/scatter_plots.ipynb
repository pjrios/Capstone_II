# Scatter Plots

input: merged.csv

output: scatter plots

```python
input_filename = "data/mergedData.csv"
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv(input_filename)

# Get the name of the last column (assuming it's the x-axis variable)
x_axis_col = df.columns[-1]

# Plot each column against the x-axis variable
for i, col in enumerate(df.columns[:-1]):
    # Create the plot
    fig, ax = plt.subplots()
    ax.scatter(df[x_axis_col], df[col], c='g', alpha=0.5, marker=r'$\clubsuit$')

    # Set axis titles and chart title
    ax.set_xlabel(x_axis_col, fontsize=12)
    ax.set_ylabel(col, fontsize=12)
    ax.set_title(f"{col} vs {x_axis_col}", fontsize=16, fontweight='bold')

    # Set font styles
    ax.tick_params(axis='both', which='major', labelsize=10)
    ax.yaxis.set_major_locator(plt.MaxNLocator(5))

    # Show the plot
    plt.show()
```
